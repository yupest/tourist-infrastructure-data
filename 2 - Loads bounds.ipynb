```python
def yandex_scraping_func(names):
    driver = webdriver.Chrome()
    for name in tqdm.notebook.tqdm(names, total = len(names)):
        get_polygons(name, driver)
#         get_polygons(name, driver)
        
def get_polygons(query, driver, v = 'default'):
    if v == 'default':
        driver.get('https://yandex.ru/maps')
    else:
        driver.get(query)
        time.sleep(1)
        query = driver.find_element_by_class_name('card-title-view__title').text
        new_queries.append(query)
    try:
        driver.find_element_by_class_name('input__control').send_keys('"'+query+'"')
        time.sleep(3)
        btn = driver.find_elements_by_class_name('small-search-form-view__button')[0]
        driver.execute_script("arguments[0].click();", btn.find_element_by_class_name('button'))
        
        time.sleep(5)
        address = driver.find_element_by_class_name('toponym-card-title-view__description').text
        coords = driver.find_element_by_class_name('toponym-card-title-view__coords-badge').text
        items.append({'place': query, 'coords':coords, 'address':address})
        
    except:
        errors.append(query)
    time.sleep(2)
```
